```python
NBA_datafile = "NBA Dataset.csv"
```

```python
import pandas as pd
import numpy as np
```

```python
#Fetching for the csv file
data = pd.read_csv(NBA_datafile)
data.head(5)
```

```python
# use filter which will create a new copy by default
new = data.filter(['TEAM','3P%'], axis=1)
```

```python
#sort the recurring states & checkins
sorted = new.sort_values(by=['TEAM', '3P%'])
```

```python
ax = sorted['TEAM'].value_counts().plot(kind='barh', figsize=(10,7),
                                        color="#22e325", fontsize=13);
ax.set_alpha(0.8)
ax.set_title("Teams with the most 3 point percentage made", fontsize=18)
ax.set_xlabel("3P%", fontsize=18);
ax.set_xticks([])

# create a list to collect the plt.patches data
totals = []

# find the values and append to list
for i in ax.patches:
    totals.append(i.get_width())

# set individual bar lables using above list
total = sum(totals)

# set individual bar lables using above list
for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width()+.3, i.get_y()+.38, \
            str(round((i.get_width()/total)*100, 2))+'%', fontsize=15,
color='dimgrey')

# invert for largest on top 
ax.invert_yaxis()
```

```python
players = data[['FULL NAME','POS','PPG']]
players.head(5)
```

```python
player_ppg = players.sort_values(ascending=False)[:5]
player_ppg = players.unstack().plot(kind='bar',stacked=True,title="Top 5 PPG by Player",figsize=(9, 7))
player_ppg.set_xlabel("FULL NAME")
player_ppg.set_ylabel("PPG")
player_ppg.legend(["Total","PG","F","SF","C","SG"], loc=9,ncol=4)
```
